```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../Data/labels.csv')
print 'Total number of data points: ' + str(data.shape[0])
data.head(5)
```

```python
print 'Number of unique labels: ' + str(len(data.tags.unique()))
print ''
unique_labels = data.tags.unique()
print data.tags.unique()
```

```python
unique_words = dict()
index = 0
for labels in unique_labels:
    words = labels.split(' ')
    for word in words:
        if word not in unique_words:
            unique_words[word] = index
            index += 1

print unique_words
```

```python
features = np.zeros((data.shape[0],len(unique_words)))
for i in range(data.shape[0]):
    words = data['tags'][i].split(' ')
    for word in words:
        features[i,unique_words[word]] = 1

for i in range(len(unique_words)):
    data[unique_words.keys()[unique_words.values().index(i)]] = features[:,i].tolist()
    
data.head(5)
```

Saving features for training.
